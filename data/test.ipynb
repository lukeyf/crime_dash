```python
import pandas as pd

codes = pd.read_csv('us-state-code.csv')
```

```python
codes.to_dict()
```

```python
df = pd.read_csv('processed_communities.csv')
df['state_abbrev'] = [str(codes.query('stname == @df.iloc[@i]["state"]')['stusps'].values[0]) for i in range(len(df))]
```

```python
df.to_csv('processed_communities.csv')
```

```python
df.groupby('state_abbrev').agg('mean').reset_index()
```
